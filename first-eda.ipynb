```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Dataset

```python
df = pd.read_csv("bearing_data.csv")

print(f"shape: {df.shape}")
print(f"columns: {df.columns}")

print(f"null values count:\n{df.isnull().sum()}")
```

### Plotting

```python
fig1, axs1 = plt.subplots(ncols=3, figsize=(10,5))
sns.histplot(data=df, x="Fr", ax=axs1[0])
sns.histplot(data=df, x="n", ax=axs1[1])
sns.histplot(data=df, x="Lifetime", ax=axs1[2])
plt.show()
```

```python
fig2, axs2 = plt.subplots(3, sharex=True, figsize=(15, 15))
axs2[0].plot(df['n'])
axs2[0].set_title('n')
axs2[1].plot(df['Fr'])
axs2[1].set_title('Fr')
axs2[2].plot(df['Lifetime'])
axs2[2].set_title('Lifetime')
```

```python
print(df.describe())
```

### Correlations

```python
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

### Variable relationships

sns.pairplot(data=df)
plt.show()
