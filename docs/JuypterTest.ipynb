```python
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
quantity = 1000

features1 = [
    "x",
    "y",
    "z"
]

df1 = pd.DataFrame(columns=features1)

df1['x'] = [random.randint(1, 6) for i in range(quantity)]

df1['y'] = [random.randint(0, 100) for i in range(quantity)]

df1['z'] = [random.randint(100, 1000) for i in range (quantity)]

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
ax.scatter3D(df1['x'], df1['y'], df1['z'], color = "green")
plt.show()
```

```python
features2 = [
    "year",
    "inflation_rate"
]
df2 = pd.DataFrame(columns=features2)

df2['year'] = [i for i in range(1945, 2023)]
df2['inflation_rate'] = [(random.randint(0, 100) / 10) - 2 for i in df2['year']]

plt.plot(df2['year'], df2['inflation_rate'])
plt.show()

plt.scatter(df2['year'], df2['inflation_rate'])
plt.show()

```
